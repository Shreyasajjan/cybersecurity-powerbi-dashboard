```python
attack_counts = df["Attack Type"].value_counts()

print(attack_counts)
```

```python
plt.figure(figsize=(10,6))
attack_counts.plot(kind="bar")

plt.title("Most Common Cyber Attacks")
plt.xlabel("Attack Type")
plt.ylabel("Count")

plt.show()
```

```python
category_counts = df["Category"].value_counts()

plt.figure(figsize=(8,8))
category_counts.plot(kind="pie", autopct='%1.1f%%')

plt.title("Security Category Distribution")
plt.ylabel("")

plt.show()
```

```python
target_counts = df["Target Type"].value_counts()

plt.figure(figsize=(10,6))
target_counts.plot(kind="bar", color="orange")

plt.title("Most Targeted Systems")
plt.xlabel("Target Type")
plt.ylabel("Count")

plt.show()
```

```python
impact_counts = df["Impact"].value_counts()

plt.figure(figsize=(10,6))
impact_counts.plot(kind="bar", color="red")

plt.title("Impact of Cyber Attacks")
plt.xlabel("Impact")
plt.ylabel("Count")

plt.show()
```
